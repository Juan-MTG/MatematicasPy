### EJERCICIO 1

#### PRIMER SISTEMA

Este sistema de ecuaciones no tiene solución ya que las rectas nunca se intersectan sino que son paralelas

```python
x = np.linspace(-6,6)
lines = plt.plot(x,(6-3*x)/2,x,(-9-6*x)/4)
plt.grid(True)
```

#### SEGUNDO SISTEMA

Este sistema de ecuaciones si tiene solución ya que las rectas si se intersectan

```python
lines = plt.plot(x,(3-x),x,(1-x)/-4)
plt.grid(True)
```

#### TERCER SISTEMA

Este sistema de ecuaciones si tiene solución ya que los planos de las ecuaciones, se intersectan

```python
y = np.linspace(-10, 6)
x2, y2 = np.meshgrid(x, y)
z1 = (3-2*x2+y2)/2
z2 = (-2*x2-3*y2)/-1
z3 = (-12-x2-2*y2)/2
fig = plt.figure()
ax = fig.add_subplot(projection="3d") 
ax.plot_surface(x2, y2, z1)
ax.plot_surface(x2, y2, z2)
ax.plot_surface(x2, y2, z3)
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_zlabel("z")
ax.set_title("SISTEMA DE ECUACIONES 3")
ax.view_init(elev=0, azim=10)

plt.show()
```

### EJERCICIO 2

Para encontrar la ecuación de la recta tangente en el punto (1,4) de la parabola, debemos derivar la ecuación de la parábola ya que esta nos da el valor de la pendiente de la recta tangente, una vez obtenido, con base en la ecuación general de la recta que es y=mx + b despejamos b y obtenemos que la recta tangente en el punto (1,4) es y = 2x + 2

```python
lines = plt.plot(x,(x**2)+3,x,2*x+2)
plt.plot(1,4,'o',color = 'red')
plt.grid(True)
plt.ylim(0,10)
plt.show()
```

### EJERCICIO 3

Para encontrar la ecuación de la recta tangente que intersecte por el punto (0.5,0) y en la parábola, debemos sustituir el valor del punto en la ecuación de la recta y luego despejar "b". Posterior a eso y teniendo el valor de "y", éste mismo valor lo sustituimos en la ecuación de la parabola, generandonos asi una nueva ecuación en la cual debemos obtener su discriminante que nos dará una ecuación cuadrática donde despejamos el valor de la pendiente, dandonos dos valores que son los valores de las dos tangentes que cortan por el mismo punto y sobre la misma parábola pero en sus dos lados, hecho esto podemo obtener por fin la ecuación de la tangente ya que obtuvimos su pendiente. Por último para obtener los puntos donde se intersectan las dos posibles tangentes con la parábola, simplemente igualamos la ecuación de la parabola con la de la tangente deseada. Obteniendo que los dos puntos donde se intersecta con cada tangente son (-1.30275 , 4.69715) y (2.30275,8.30256).

```python
lines = plt.plot(x,(x**2)+3,x,4.6055*x -2.30275,x,-2.6055*x + 1.30275)
plt.plot(0.5,0,'o',color = 'blue')
plt.plot(-1.30275 , 4.69715,'o',color = 'red')
plt.plot(2.30275,8.30256,'o',color = 'red')
plt.grid(True)
plt.ylim(-1,30)
plt.xlim(-10,10)
plt.show()
```

### EJERCICIO 2

```python
print("----------A---------")
A2 = np.array([[3,0],[-1,5]])
print(A2)
print("----------B---------")
B2 = np.array([[4,-2,1],[0,2,3]])
print(B2)
```

###### Si las matrices A=(aij) y B=(bij) tienen la misma dimensión, la matriz suma es:
    A+B=(aij+bij).
Debido a que nuestras matrices no son de la misma dimensión no se puede realizar la operación de suma entre
éstas dos, sin embargo la operación de multiplicación de una constante (3) por la matriz B si se puede realizar
ya que la multiplicación de un escalar por una matriz es el escalar multiplicando a cada uno de los elementos
de la matriz.

###### INCISO 1

```python
print("----------A + 3B---------")
print(3*B2)
C = A2 + (3*B2)
```

###### INCISO 2

###### La multiplicación de matrices se logró
Ya que A2 es de 2x2 y B2 es de 2x3 teniendo así la primera el mismo numero de columnas que el número de filas de la segunda, resultando una matriz D de 2x3

```python
print("----------AB---------")
D = A2@B2
print(D)
```

###### INCISO 3

###### POTENCIA EN MATRICES 
La potencia de una matriz es igual a multiplicar esa matriz consigo misma las veces "n" que se indiquen. Sin embargo el declarar MATRIZ \**n elevara cada elemendo individual de la matriz a la "n", por lo que hablando matemáticamente, no sería el resultado correcto de la elevación de una matriz a la potencia "n".

Dicho lo anterior entonces debemos ya sea usar el operador @ para la multiplicación de matrices el numero de veces que se indique en la potencia o usar la función que nos da numpy para hacer esto mismo, la cual se llama:
**np.linalg.matrix_power(matriz,potencia)**

```python
print("----------A^3---------")
print("----------ELEVANDO INDIVIDUALMENTE CADA ELEMENTO DE LA MATRIZ AL CUBO---------")
print(A2**3)
print("----------ELEVANDO LA MATRIZ AL CUBO---------")
print(A2@A2@A2)
print("----------ELEVANDO LA MATRIZ AL CUBO CON LA FUNCIÓN DE NUMPY---------")
print(np.linalg.matrix_power(A2,3))
```

###### INCISO 4

La multiplicación matricial y la potencia de matrices están restringidas a matrices cuadradas debido a las propiedades y definiciones específicas de estas operaciones. En pocas palabras la potencia de matrices solo puede ser realizada en matrices cuadras, cosa que no sucede en la matriz B ya que esta es de 2x3. Puede realizarse la elevación individual de sus elementos a la "n" potencia, pero no en definición.

```python
print("----------B**2---------")
print("----------ELEVANDO INDIVIDUALMENTE CADA ELEMENTO DE LA MATRIZ AL CUADRADO---------")
print(B2**2)
print("----------ELEVANDO LA MATRIZ AL CUBO---------")
print(B2@B2@B2)
print("----------ELEVANDO LA MATRIZ AL CUBO CON LA FUNCIÓN DE NUMPY---------")
print(np.linalg.matrix_power(B2,2))
```

###### INCISO 5

Aquí sucede algo parecido al inciso 2 ya que el multiplicar la matriz A por un escalar no afecta la estructura de la misma, por lo que le permite después poder hacer la multiplicación de matrices con la matriz B

###### SISTEMA DE ECUACIÓN 1

El sistema de ecuaciones planteado no tiene una solución, ya que la rectas de cada ecuación nunca se intersectan, sino que son paralelas.

```python
#INSERTAMOS EN UN ARREGLO LAS ECUACIONES
eq1 = np.array([[3,2],[6,4]])
print(eq1)
#INSERTAMOS EN UN ARREGLO LA PARTE INDEPENDIENTE DE LAS ECUACIONES
ind1 = np.array([6,-9])
print(ind1)
#CON NUMPY Y SU METODO linalg.solve(ECUACIONES,INDEPENDIENTES)
#RESOLVEMOS EL SISTEMA DE ECUACIONES
print("----------------SOLUCIÓN---------------")
sol1 = np.linalg.solve(eq1,ind1)
print(sol1)
```

###### SISTEMA DE ECUACIÓN 2

Debido a que este sistema de ecuaciones si tiene solución es entonces que no genera error nuestra función: **np.linalg.solve(eq2,ind2)**

```python
#INSERTAMOS EN UN ARREGLO LAS ECUACIONES
eq2 = np.array([[1,1],[1,-4]])
print(eq2)
#INSERTAMOS EN UN ARREGLO LA PARTE INDEPENDIENTE DE LAS ECUACIONES
ind2 = np.array([3,1])
print(ind2)
#CON NUMPY Y SU METODO linalg.solve(ECUACIONES,INDEPENDIENTES)
#RESOLVEMOS EL SISTEMA DE ECUACIONES
print("----------------SOLUCIÓN---------------")
sol2 = np.linalg.solve(eq2,ind2)
print(f"x= {sol2[0]}    |   y = {sol2[1]}")
print("----------------COMPROBACIÓN ECUACION 1---------------")
print((eq2[0,0]*sol2[0] + eq2[0,1]*sol2[1]) == ind2[0])
print("----------------COMPROBACIÓN ECUACION 2---------------")
print((eq2[1,0]*sol2[0] + eq2[1,1]*sol2[1]) == ind2[1])
```

###### SISTEMA DE ECUACIÓN 3

```python
#INSERTAMOS EN UN ARREGLO LAS ECUACIONES
eq3 = np.array([[2,-1,2],[2,3,-1],[1,2,2]])
print(eq3)
#INSERTAMOS EN UN ARREGLO LA PARTE INDEPENDIENTE DE LAS ECUACIONES
ind3 = np.array([3,0,-12])
print(ind3)
#CON NUMPY Y SU METODO linalg.solve(ECUACIONES,INDEPENDIENTES)
#RESOLVEMOS EL SISTEMA DE ECUACIONES
print("----------------SOLUCIÓN---------------")
sol3 = np.linalg.solve(eq3,ind3)
print(f"x= {sol3[0]}    |   y = {sol3[1]}    |   z = {sol3[2]}")
print("----------------COMPROBACIÓN ECUACION 1---------------")
print((eq3[0,0]*sol3[0] + eq3[0,1]*sol3[1] + eq3[0,2]*sol3[2]) == ind3[0])
print("----------------COMPROBACIÓN ECUACION 2---------------")
#SUAMOS LA FUNCION trunch DE math PARA QUITAR LOS DECIMALES, YA QUE NOS DABA UN NÚMERO QUE ES PRACTICAMENTE 0, DONDE
#PODEMOS DESPRECIAR LOS DECIMALES
print(math.trunc(eq3[1,0]*sol3[0] + eq3[1,1]*sol3[1] + eq3[1,2]*sol3[2]) == ind3[1])
print("----------------COMPROBACIÓN ECUACION 3---------------")
print((eq3[2,0]*sol3[0] + eq3[2,1]*sol3[1] + eq3[2,2]*sol3[2]) == ind3[2])
```
